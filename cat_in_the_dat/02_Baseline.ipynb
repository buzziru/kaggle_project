# Module

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
```

# Data

```python
data_path = "./data/"

train = pd.read_csv(data_path + 'train.csv', index_col='id')
test = pd.read_csv(data_path + 'test.csv', index_col='id')
submission = pd.read_csv(data_path + 'sample_submission.csv', index_col='id')

train.head()
```

```python
train.shape, test.shape, submission.shape
```

```python
all_data = pd.concat([train, test], axis=0)
all_data.shape
```

```python
all_data.drop('target', axis=1, inplace=True)
all_data.shape
```

```python
all_data.head()
```

# Feature Engineering

```python
# Encoding
encoder = OneHotEncoder()
all_data_encoded = encoder.fit_transform(all_data)
```

```python
num_train = len(train)
x_train = all_data_encoded[:num_train]
x_test = all_data_encoded[num_train:]

y = train['target']
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_train, y, test_size=0.1, 
                                                  stratify=y, random_state=2025)
```

# Model

```python
logistic_model = LogisticRegression(max_iter=1000, random_state=2025)
logistic_model.fit(x_train, y_train)
```

```python
y_val_preds = logistic_model.predict_proba(x_val)[:, 1]

roc_auc = roc_auc_score(y_val, y_val_preds)
print(f"ROC AUC: {roc_auc:.4f}")
```
